```python
import pandas as pd
import datetime as dt
```

```python
#Create a DataFrame for the 202301-citibike-tripdata
csv_path = "Resource/202301-citibike-tripdata.csv"
df = pd.read_csv(csv_path)
df.head()
```

```python
#rechecking the number of riders in the new dataset
df2.count()
```

```python
#Clearing duplicates by ride_id if existing
df2.drop_duplicates(subset='ride_id')
```

```python
#Rechecking numbers
df2.count()
```

```python
#Uploading dataset for 202212-citibike-tripdata.csv
csv_path2 = "Resource/202212-citibike-tripdata.csv"
df3 = pd.read_csv(csv_path2)
df3.head()
```

```python
#Checking count of initial 202212 spreadsheet
df3.count()
```

```python
#Confirming that datatypes match with 202301
df3.dtypes
```
